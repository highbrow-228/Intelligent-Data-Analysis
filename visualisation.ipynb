## <b><span style='color:#F1C40F'>|</span> Візуалізація</b>

### <b><span style='color:#39dd73'>|</span> Розподіл за віковими групами</b>

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10))
sns.countplot(x='AGE', data=data, hue='AGE')
plt.title('Розподіл за віковими групами')
plt.xlabel('Вікова група')
plt.ylabel('Частота')
plt.show()
```

### <b><span style='color:#39dd73'>|</span> Вплив віку на результати страхування</b>

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='AGE', hue='OUTCOME', data=data)
plt.title('Вплив віку на результати страхування.')
plt.xlabel('Вікова група')
plt.ylabel('Частота')
plt.legend(title='(1=Отримав, 0=Не отримав)')
plt.show()
```

### <b><span style='color:#39dd73'>|</span> Вплив статі на результати страхування</b>

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='GENDER', hue='OUTCOME', data=data)
plt.title('Вплив статі на результати страхування')
plt.xlabel('Стать')
plt.ylabel('Частота')
plt.legend(title='(1=Отримав, 0=Не отримав)')
plt.show()
```

### <b><span style='color:#39dd73'>|</span> Розподіл досвіду водіння</b>

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='DRIVING_EXPERIENCE', data=data, order=['0-9y', '10-19y', '20-29y', '30y+'], hue='OUTCOME')
plt.title('Розподіл досвіду водіння')
plt.xlabel('Стаж водіння (років)')
plt.ylabel('Частота')
plt.legend(title='(1=Отримав, 0=Не отримав)')
plt.show()
```

### <b><span style='color:#39dd73'>|</span> Розподіл власності на транспортні засоби</b>

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='VEHICLE_OWNERSHIP', data=data, hue='OUTCOME', legend=False)
plt.title('Розподіл власності на транспортні засоби')
plt.xlabel('Володіння транспортним засобом (1=Так, 0=Ні)')
plt.ylabel('Частота')
plt.show()
```

### <b><span style='color:#39dd73'>|</span> Розподіл за типами транспортних засобів</b>

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='VEHICLE_TYPE', data=data, hue='OUTCOME', legend=False)
plt.title('Розподіл за типами транспортних засобів')
plt.xlabel('Тип транспортного засобу')
plt.ylabel('Частота')
plt.show()
```

### <b><span style='color:#39dd73'>|</span> Вплив сімейного статусу на отримання автоцивілки</b>

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='MARRIED', data=data, hue='OUTCOME')
plt.title('Вплив сімейного статусу на отримання автоцивілки')
plt.xlabel('Сімейний статус (1=Одружений, 0=Не одружений)')
plt.ylabel('Кількість')
plt.legend(title='Результат (1=Отримав, 0=Не отримав)')
plt.show()
```

### <b><span style='color:#39dd73'>|</span> Вплив рівня освіти на отримання автоцивілки</b>

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='EDUCATION', data=data, hue='OUTCOME', order=['none', 'high school', 'university'])
plt.title('Вплив рівня освіти на отримання автоцивілки')
plt.xlabel('Рівень освіти')
plt.ylabel('Кількість')
plt.legend(title='Результат (1=Отримав, 0=Не отримав)')
plt.show()
```
