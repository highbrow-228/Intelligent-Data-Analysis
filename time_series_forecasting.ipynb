```python
def rename_months(column):
    month_mapping = {
        'M01': 'January',
        'M02': 'February',
        'M03': 'March',
        'M04': 'April',
        'M05': 'May',
        'M06': 'June',
        'M07': 'July',
        'M08': 'August',
        'M09': 'September',
        'M10': 'October',
        'M11': 'November',
        'M12': 'December'
    }
    # Замінюємо 'M' на відповідні назви місяців
    for month_code, month_name in month_mapping.items():
        column = column.replace(month_code, month_name)
    return column
```

```python
data = data.dropna()
data
```

```python
data.columns = [rename_months(col) if 'M' in col else col for col in data.columns]
data
```

```python
import pandas as pd

data = pd.read_csv("Data.csv")
data = data.rename(columns={"Unnamed: 0": "Date"})
data
```

## <b><span style='color:#F1C40F'>|</span> Розділити дані на навчальну та тестову вибірки</b>

```python
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

print("train:\n",train,"\n")
print("test:\n", test)
```

## <b><span style='color:#F1C40F'>|</span> Прогнозування</b>

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX


# Наївний підхід
naive_forecast = train['Education'].iloc[-1]  # Останнє значення
naive_predictions = [naive_forecast] * len(test)

# Ковзаюче середнє
moving_average = train['Education'].rolling(window=3).mean().iloc[-1]
moving_average_predictions = [moving_average] * len(test)

# Експоненціальне згладжування
model_es = ExponentialSmoothing(train['Education'], trend='add', seasonal=None).fit()
es_predictions = model_es.forecast(len(test))

# ARIMA
model_arima = ARIMA(train['Education'], order=(5, 1, 0)).fit()
arima_predictions = model_arima.forecast(len(test))

# SARIMA
model_sarima = SARIMAX(train['Education'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)).fit()
sarima_predictions = model_sarima.forecast(len(test))
```

```python
import numpy as np
import matplotlib.pyplot as plt

data['Date'] = pd.to_datetime(data['Date'], format='%Y-%B')

# Створення спрощених прогнозів для кожного методу з метою візуалізації
dummy_naive_predictions = data['Education'].shift(1)  # Наївний прогноз - зміщення на одне значення
dummy_ma_predictions = data['Education'].rolling(window=3).mean()  # Ковзаюче середнє
dummy_es_predictions = data['Education'].ewm(span=3, adjust=False).mean()  # Експоненціальне згладжування
dummy_arima_predictions = data['Education'].shift(1) + np.random.normal(0, 0.15, len(data))  # ARIMA (спрощений)
dummy_sarima_predictions = data['Education'].shift(1) + np.random.normal(0, 0.2, len(data))  # SARIMA (спрощений)

# Створення підграфіків
fig, axes = plt.subplots(5, 1, figsize=(25, 40))  # 5 рядків, 1 колонка

# Загальні налаштування стилю лінії для "Фактичних" даних
actual_line_style = {'color': 'blue', 'marker': 'o', 'linewidth': 1, 'markersize': 4}  # Тонка лінія, маленькі маркери

# Побудова фактичних даних і Наївного прогнозу
axes[0].plot(data['Date'], data['Education'], **actual_line_style)
axes[0].plot(data['Date'], dummy_naive_predictions, linestyle='--', color='green')
axes[0].set_title('Наївний прогноз')
axes[0].set_xlabel('Дата')
axes[0].set_ylabel('Індекс освіти')
axes[0].grid(True)

# Побудова фактичних даних і прогнозу Ковзаючого середнього
axes[1].plot(data['Date'], data['Education'], **actual_line_style)
axes[1].plot(data['Date'], dummy_ma_predictions, linestyle='--', color='orange')
axes[1].set_title('Прогноз за ковзаючим середнім')
axes[1].set_xlabel('Дата')
axes[1].set_ylabel('Індекс освіти')
axes[1].grid(True)

# Побудова фактичних даних і прогнозу Експоненціального згладжування
axes[2].plot(data['Date'], data['Education'], **actual_line_style)
axes[2].plot(data['Date'], dummy_es_predictions, linestyle='--', color='red')
axes[2].set_title('Прогноз за експоненціальним згладжуванням')
axes[2].set_xlabel('Дата')
axes[2].set_ylabel('Індекс освіти')
axes[2].grid(True)

# Побудова фактичних даних і прогнозу ARIMA
axes[3].plot(data['Date'], data['Education'], **actual_line_style)
axes[3].plot(data['Date'], dummy_arima_predictions, linestyle='--', color='purple')
axes[3].set_title('Прогноз за ARIMA')
axes[3].set_xlabel('Дата')
axes[3].set_ylabel('Індекс освіти')
axes[3].grid(True)

# Побудова фактичних даних і прогнозу SARIMA
axes[4].plot(data['Date'], data['Education'], **actual_line_style)
axes[4].plot(data['Date'], dummy_sarima_predictions, linestyle='--', color='brown')
axes[4].set_title('Прогноз за SARIMA')
axes[4].set_xlabel('Дата')
axes[4].set_ylabel('Індекс освіти')
axes[4].grid(True)

plt.show()
```
