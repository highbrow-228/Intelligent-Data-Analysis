## <b><span style='color:#2692c9'>|</span> Кореляції</b>

```python
import matplotlib.pyplot as plt
import seaborn as sns

pearson_correlation_matrix = data.corr(method='pearson')
pearson_correlation_matrix = data.corr().round(2)
plt.figure(figsize=(25, 10))
sns.heatmap(
    pearson_correlation_matrix, 
    annot=True,
    linecolor='black',
    linewidth=1,
)
plt.title("Кореляція Пірсона")
plt.show()
```

```python
spearman_correlation_matrix = data.corr(method='spearman')
spearman_correlation_matrix = data.corr().round(2)

plt.figure(figsize=(25, 10))
sns.heatmap(
    spearman_correlation_matrix, 
    annot=True,
    linecolor='black',
    linewidth=1,
)
plt.title("Кореляція Спірмена")
plt.show()
```

```python
kendall_correlation_matrix = data.corr(method='kendall')
kendall_correlation_matrix = data.corr().round(2)
plt.figure(figsize=(25, 10))
sns.heatmap(
    kendall_correlation_matrix, 
    annot=True,
    linecolor='black',
    linewidth=1,
)
plt.title("Кореляція Кендалла")
plt.show()
```

## <b><span style='color:#2692c9'>|</span> Графіки</b>

### <b><span style='color:#1dcd5b'>|</span> Розподіл за доходом</b>

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Income'], kde=True)
plt.title('Розподіл за доходом')
plt.xlabel('Дохід')
plt.ylabel('Частота')
plt.show()
```

### <b><span style='color:#1dcd5b'>|</span> Розподіл за віком</b>

```python
# Plot distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True)
plt.title('Розподіл за віком')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

### <b><span style='color:#1dcd5b'>|</span> Середні витрати на категорії продуктів</b>

```python
categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
mean_expenses = data[categories].mean()

plt.figure(figsize=(10, 6))
mean_expenses.plot(kind='bar')
plt.title('Середні витрати на категорії продуктів')
plt.xlabel('Категорії продуктів')
plt.ylabel('Середні витрати')
plt.grid(True)
plt.show()
```

### <b><span style='color:#1dcd5b'>|</span> Витрати на продукти за віковими групами</b>

```python
data['Age_Group'] = pd.cut(data['Age'], bins=[18, 30, 40, 50, 60, 100], labels=['18-30', '31-40', '41-50', '51-60', '60+'])
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='MntTotal', data=data)
plt.title('Витрати на продукти за віковими групами')
plt.xlabel('Вікова група')
plt.ylabel('Витрати на продукти')
plt.grid(True)
plt.show()
```

### <b><span style='color:#1dcd5b'>|</span> Витрати в залежності від кількості дітей</b>

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Kidhome', y='MntTotal', data=data)
plt.title('Витрати в залежності від кількості дітей')
plt.xlabel('Кількість дітей (Kidhome)')
plt.ylabel('Витрати на продукти')
plt.grid(True)
plt.show()
```
